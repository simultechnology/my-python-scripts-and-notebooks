```python
import numpy as np 
import pandas as pd 
```

```python
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../data/titanic.csv")
```

```python
data.head()
```

```python
data.describe()
```

```python
data.head(10)
```

```python
df = data[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Survived']]
```

```python
df['Sex'] = df['Sex'].map({'male': 0, 'female':1})
```

```python
df.head()
```

```python
# process 'Cabin'
char_cabin = data['Cabin'].astype(str)
```

```python
new_cabin = np.array([cabin[0] for cabin in char_cabin])
```

```python
data['Cabin'] = pd.Categorical(char_cabin)
```

```python
# process 'Age'
mean_age = data.Age.mean()
mean_age
```

```python
new_age_var = np.where(data.Age.isnull(), mean_age, data.Age)
```

```python
data.Age = new_age_var
```

```python
data.head(10)
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
```

```python
encoded_sex = label_encoder.fit_transform(data.Sex)
```

```python
encoded_class = label_encoder.fit_transform(data.Pclass)
encoded_cabin = label_encoder.fit_transform(data.Cabin)
```

```python
train_features = pd.DataFrame([encoded_class, encoded_cabin, encoded_sex, data.Age]).T
```

```python
# Building up the Decision Tree model
from sklearn import tree
dtree = tree.DecisionTreeClassifier(criterion='entropy')
```

```python
dtree
```

```python
dtree.fit(X=train_features, y=data.Survived)
```

```python
preds = dtree.predict(X=train_features)
```

```python
dtree.score(X=train_features, y = data.Survived)
```

```python
from sklearn import metrics
metrics.confusion_matrix(y_true= data.Survived, y_pred=preds)
```

```python
# Classification report
print(metrics.classification_report(y_true=data.Survived, y_pred=preds))
```
